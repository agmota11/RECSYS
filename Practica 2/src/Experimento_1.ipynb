```python
import pandas as pd
from datetime import datetime
import numpy as np

# Load train and test CSV files into DataFrames
train_df = pd.read_csv('../data/train_reviews.csv')
test_df = pd.read_csv('../data/test_reviews.csv')

train_df.sample(10)
```

```python
test_df.sample(10)
```

```python
rating_distribution = train_df['stars'].value_counts().sort_index()
print(rating_distribution)
```

```python
# Calcular la moda de las calificaciones en el DataFrame de entrenamiento
rating_mode = train_df['stars'].mode()[0]

# Asignar la moda como calificaci√≥n para todos los registros en el DataFrame de prueba
test_df['stars'] = rating_mode

# Exportar el resultado a un archivo CSV
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_path = f'../out/result_{timestamp}.csv'
test_df[['review_id', 'stars']].to_csv(output_path, index=False)

print(f"File saved to {output_path}")
test_df.sample(10)
```

```python
from datetime import datetime

rating_mode = train_df['stars'].mean()

test_df['stars'] = rating_mode.round(0)

timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_path = f'../out/result_{timestamp}.csv'
test_df[['review_id', 'stars']].to_csv(output_path, index=False)

print(f"File saved to {output_path}")
test_df.sample(10)
```

```python
stars_distribution = train_df['stars'].value_counts().sort_index()
print(stars_distribution)
```

## Predecir usando distribuciones probabilisticas

```python
# Normalize the rating distribution to get probabilities
rating_probs = stars_distribution / stars_distribution.sum()

# Generate ratings for the test DataFrame based on the train distribution
test_df['stars'] = np.random.choice(rating_distribution.index, size=len(test_df), p=rating_probs)

# Prepare the output file path with a timestamp
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_path = f'../out/result_{timestamp}.csv'

# Export the result to an Excel file
test_df[['review_id', 'stars']].to_csv(output_path, index=False)

print(f"File saved to {output_path}")
test_df.sample(10)
```

## Random [3.0, 4.0, 5.0]

```python
test_df['stars'] = np.random.choice([3.0, 4.0, 5.0], size=len(test_df), p=[0.3, 0.5, 0.2])

timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_path = f'../out/result_{timestamp}.csv'
test_df[['review_id', 'stars']].to_csv(output_path, index=False)

print(f"File saved to {output_path}")
test_df.sample(5)
```
