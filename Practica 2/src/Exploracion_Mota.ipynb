```python
from sklearn.preprocessing import StandardScaler

def process_users(df_users, scalers):
    df_users = df_users.copy()
    df_users['elite'] = df_users['elite'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
    df_users['friends'] = df_users['friends'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
    df_users['yelping_since'] = pd.to_datetime(df_users['yelping_since'])
    df_users['yelping_since'] = (datetime.now() - df_users['yelping_since']).dt.days
    df_users['yelping_since'] = df_users['yelping_since'].apply(lambda x: x // 30)

    # drop columns
    df_users.drop(columns=['name'], inplace=True)   

    # StandardScaler all columns except 'user_id'
    cols = list(df_users.columns)
    cols.remove('user_id')
    for col in cols:
        scaler = StandardScaler()
        df_users[col] = scaler.fit_transform(df_users[[col]])
        scalers[col] = scaler

    # Add users prefix to column names
    df_users.columns = ['users_' + col if col != 'user_id' else col for col in df_users.columns]

    return df_users 

std_scalers = {}
df_processed_users = process_users(usuarios, std_scalers)
df_processed_users.sample(10)
```

```python
# Sample category grouping — you should expand this based on your domain knowledge
CATEGORY_GROUPS = {
    "Food & Dining": [
        "Restaurants", "Bakeries", "Fast Food", "Buffets", "Bars", "Cafes", "Coffee & Tea",
        "Pizza", "Burgers", "Breakfast & Brunch", "Seafood", "Steakhouses", "Sushi Bars",
        "Hot Dogs", "Hot Pot", "Ice Cream & Frozen Yogurt", "Juice Bars & Smoothies"
    ],
    "Health & Medical": [
        "Dentists", "Doctors", "Chiropractors", "Dermatologists", "Hospitals", "Optometrists",
        "Psychiatrists", "Psychologists", "Urgent Care", "Pediatricians", "Acupuncture",
        "Medical Centers", "Medical Spas"
    ],
    "Beauty & Spas": [
        "Hair Salons", "Nail Salons", "Massage", "Day Spas", "Hair Removal", "Skin Care",
        "Waxing", "Makeup Artists", "Cosmetic Surgeons"
    ],
    "Automotive": [
        "Auto Repair", "Car Dealers", "Car Wash", "Auto Glass Services", "Auto Parts & Supplies",
        "Oil Change Stations", "Tires"
    ],
    "Education": [
        "Colleges & Universities", "Tutoring Centers", "Educational Services", "Vocational & Technical School"
    ],
    "Entertainment": [
        "Amusement Parks", "Arcades", "Bowling", "Dance Clubs", "Comedy Clubs", "Zoos", "Museums"
    ]
}

# Reverse map: fine category -> group
FINE_TO_GROUP = {cat: group for group, cats in CATEGORY_GROUPS.items() for cat in cats}
```

```python
def group_categories(category_string):
    """Map fine-grained categories into high-level groups."""
    seen_groups = set()
    for cat in category_string.split(','):
        cat = cat.strip()
        if cat in FINE_TO_GROUP:
            seen_groups.add(FINE_TO_GROUP[cat])
    return list(seen_groups)

def one_hot_encode_groups(df, category_col='buss_categories'):
    """One-hot encode grouped categories into broader buckets."""
    df[category_col].fillna('', inplace=True)
    df['grouped_categories'] = df[category_col].apply(group_categories)
    
    # Create dummy variables from the grouped categories
    category_dummies = df['grouped_categories'].apply(lambda x: pd.Series(1, index=x)).fillna(0).astype(int)
    df = pd.concat([df, category_dummies], axis=1)
    return df
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

def process_businesses(df_businesses, scalers):
    df_businesses = df_businesses.copy()

    # add buss_ prefix to all columns except 'business_id'
    df_businesses.columns = ['buss_' + col if col != 'business_id' else col for col in df_businesses.columns]

    # one hot encoder for state and ordinal encode for zipcode
    ohe = OneHotEncoder(sparse_output=False)  # Ensure dense output
    ohe_states = ohe.fit_transform(df_businesses[['buss_state']])
    ohe_states = pd.DataFrame(ohe_states, columns=ohe.get_feature_names_out(['buss_state']), index=df_businesses.index)
    df_businesses = pd.concat([df_businesses, ohe_states], axis=1)
    df_businesses.drop(columns=['buss_state'], inplace=True)
    scalers['buss_state'] = ohe

    oe = OrdinalEncoder()
    df_businesses['buss_postal_code'] = df_businesses['buss_postal_code'].astype(str)
    df_businesses['buss_postal_code'].fillna('unknown', inplace=True)
    df_businesses['buss_postal_code'] = oe.fit_transform(df_businesses[['buss_postal_code']])
    df_businesses['buss_postal_code'] = df_businesses['buss_postal_code'].astype(int)
    scalers['buss_postal_code'] = oe

    # one hot encode for categories
    # Optimize category encoding by creating a single DataFrame for all categories
    df_businesses = one_hot_encode_groups(df_businesses)
       
    df_businesses.drop(columns=['buss_name', 'buss_address', 'buss_city', 'buss_is_open', 'buss_attributes', 'buss_hours', 'buss_categories', 'grouped_categories'], inplace=True)   

    # StandardScaler 
    cols = ['buss_review_count', 'buss_latitude', 'buss_longitude']
    for col in cols:
        scaler = StandardScaler()
        df_businesses[col] = scaler.fit_transform(df_businesses[[col]])
        scalers[col] = scaler

    return df_businesses

df_processed_bussinesses = process_businesses(negocios, std_scalers)
df_processed_bussinesses.sample(10)
```

```python
def process_reviews(df_reviews, scalers):
    df_reviews = df_reviews.copy()

    # StandardScaler for review-related columns
    cols = ['useful', 'funny', 'cool']
    for col in cols:
        scaler = StandardScaler()
        df_reviews[col] = scaler.fit_transform(df_reviews[[col]])
        scalers[col] = scaler

    df_reviews['date'] = pd.to_datetime(df_reviews['date'])
    df_reviews['date'] = (datetime.now() - df_reviews['date']).dt.days
    df_reviews['date'] = df_reviews['date'].apply(lambda x: x // 30)

    df_reviews.drop(columns=['text'], inplace=True)  # Drop review text for simplicity

    # add review_ prefix to all columns except 'review_id'
    df_reviews.columns = ['review_' + col if col not in ['review_id', 'business_id', 'user_id'] else col for col in df_reviews.columns]

    return df_reviews

def process_reviews_test(df_reviews, scalers):
    df_reviews = df_reviews.copy()

    # StandardScaler for review-related columns
    cols = ['useful', 'funny', 'cool']
    for col in cols:
        scaler = scalers[col]
        df_reviews[col] = scaler.transform(df_reviews[[col]])

    df_reviews['date'] = pd.to_datetime(df_reviews['date'])
    df_reviews['date'] = (datetime.now() - df_reviews['date']).dt.days
    df_reviews['date'] = df_reviews['date'].apply(lambda x: x // 30)

    df_reviews.drop(columns=['text'], inplace=True)  # Drop review text for simplicity

    # add review_ prefix to all columns except 'review_id'
    df_reviews.columns = ['review_' + col if col not in ['review_id', 'business_id', 'user_id'] else col for col in df_reviews.columns]

    return df_reviews

df_processed_reviews = process_reviews(train_reviews, std_scalers)
df_processed_reviews.sample(10)
```

```python
df_processed_reviews_test = process_reviews_test(test_reviews, std_scalers)
df_processed_reviews_test.sample(10)
```
