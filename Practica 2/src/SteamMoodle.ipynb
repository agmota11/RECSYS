```python
import pandas as pd
```

```python
raw_games = pd.read_csv('steam/steam.csv', index_col='appid')
raw_descriptions = pd.read_csv('steam/steam_description_data.csv', index_col='steam_appid')
```

```python
games = raw_games.join(raw_descriptions)
```

```python
games.shape
```

```python
games.reset_index(inplace=True)
```

```python
games['content'] = games['name'].fillna("") + " " + games['detailed_description'].fillna("")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances

tfidf = TfidfVectorizer(stop_words='english', max_features=15000, ngram_range=(1, 2)).fit_transform(games['content'])
```

```python
tfidf.shape
```

```python
m = cosine_distances(tfidf)
```

```python
import numpy as np
print(f"Top 10 juegos más parecidos a {games.iloc[6433]['name']}")
idxs = sorted(enumerate(m[6433]), key=lambda x: x[1], reverse=False)
for i, score in idxs[1:11]:
    print(f"{games.iloc[i]['name']}: {score: .5f}")
```

```python
player_ids = [1225, 2684, 222, 1509]
d = tfidf[player_ids].sum(axis=0)
d = np.asarray(d).reshape(1, -1)
```

```python
d = cosine_distances(d, tfidf)[0]
```

```python
print(f"Top 10 juegos para jugador misterioso")
idxs = sorted(enumerate(d), key=lambda x: x[1], reverse=False)
for i, score in [x for x in idxs if x[0] not in player_ids]:
    print(f"{games.iloc[i]['name']}: {score: .5f}")
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim

model = SentenceTransformer('all-MiniLM-L12-v2')

corpus = list(games['content'])
embeddings = model.encode(corpus)
```

```python
mm = cos_sim(embeddings, embeddings)
```
